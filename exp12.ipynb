```python
class PVModule:
    def __init__(self, V_oc, I_sc, V_mp, I_mp):
        self.V_oc = V_oc # Open circuit voltage (V)
        self.I_sc = I_sc # Short circuit current (A)
        self.V_mp = V_mp # Voltage at maximum power point (V)
        self.I_mp = I_mp # Current at maximum power point (A)
```

```python
def calculate_output_voltage(self, irradiance, temperature):
    """
    Calculate the output voltage of the PV module based on the irradiance and temperature.
    """
    # Example model (you can replace this with more accurate models)
    V = self.V_oc - (irradiance / 1000) * 0.1 - (temperature - 25) * 0.05
    return V
```

```python
 class MPPTController:
    def __init__(self, V_ref):
        self.V_ref = V_ref
```

```python
def track_voltage(self, pv_module, irradiance, temperature):
    """
    Track the desired output voltage using a simple voltage control strategy.
    """
    V_out = pv_module.calculate_output_voltage(irradiance, temperature)
    error = self.V_ref - V_out
    return error
```

```python
# Simulation parameters
V_ref = 36 # Desired output voltage (V)
irradiance = 1000 # Irradiance (W/m^2)
temperature = 25 # Temperature (°C)
```

```python
# Create PV module and MPPT controller objects
pv_module = PVModule(V_oc=36, I_sc=8.3, V_mp=29, I_mp=7.6)
controller = MPPTController(V_ref)
```

```python
# Simulate MPPT control
error = controller.track_voltage(pv_module, irradiance, temperature)
print("Error:", error)
```
